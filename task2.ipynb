```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "C:\\Users\\SUVARCHALA\\OneDrive\\Documents\\titanic data.csv"  # Change this to your file path if running locally
titanic = pd.read_csv(url)
```

```python
# Step 1: Data Cleaning
# Handle missing values
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
titanic.drop(columns=['Cabin'], inplace=True)
titanic.dropna(inplace=True)
```

```python

# Convert categorical variables to numeric
titanic['Sex'] = titanic['Sex'].map({'male': 0, 'female': 1})
titanic['Embarked'] = titanic['Embarked'].map({'C': 0, 'Q': 1, 'S': 2})
```

```python

# Step 2: Exploratory Data Analysis (EDA)
# Summary statistics
print(titanic.describe())
```

```python
# Visualize distributions
plt.figure(figsize=(12, 6))
sns.histplot(titanic['Age'], kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Survived vs. Non-Survived
plt.figure(figsize=(12, 6))
sns.countplot(x='Survived', data=titanic, palette='viridis')
plt.title('Survival Counts')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

```

```python

# Explore relationships between variables
# Survival rate by sex
plt.figure(figsize=(12, 6))
sns.barplot(x='Sex', y='Survived', data=titanic, palette='viridis')
plt.title('Survival Rate by Sex')
plt.xlabel('Sex (0 = Male, 1 = Female)')
plt.ylabel('Survival Rate')
plt.show()
```

```python

# Survival rate by class
plt.figure(figsize=(12, 6))
sns.barplot(x='Pclass', y='Survived', data=titanic, palette='viridis')
plt.title('Survival Rate by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```

```python

# Pairplot to see relationships
sns.pairplot(titanic[['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']], hue='Survived', palette='viridis')
plt.show()
```
